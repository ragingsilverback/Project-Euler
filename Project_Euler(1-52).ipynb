**Problem 9**

---



A Pythagorean triplet is a set of three natural numbers, a < b < c, for which,

a2 + b2 = c2
For example, 32 + 42 = 9 + 16 = 25 = 52.

There exists exactly one Pythagorean triplet for which a + b + c = 1000.
Find the product abc.

```
def pythagorean_triplet(sum):
  sol = []
  for a in range(1,sum//2):
    for b in range(1,sum//2):
      for c in range(1,sum//2):
        if( a*a + b*b == c*c and a + b + c == sum):
          sol.append([a,b,c])
          
  return sol

pythagorean_triplet(1000)
```

The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

 1: 1
 3: 1,3
 6: 1,2,3,6
10: 1,2,5,10
15: 1,3,5,15
21: 1,3,7,21
28: 1,2,4,7,14,28
We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

```
import math

def triangle_number(n):
  return n*(n+1)//2

def factors(n):
  factors = 0
  sqrt  = int(math.sqrt(n))
  for i in range(1,sqrt):
    if(n%i == 0):
      factors += 2
  if(sqrt*sqrt == n):
    factors = factors - 1
  return factors

n = 1
while(factors(triangle_number(n)) < 500):
  n = n + 1

print(n)
triangle_number(n)
```

**Problem 14**


---
The following iterative sequence is defined for the set of positive integers:

n → n/2 (n is even)
n → 3n + 1 (n is odd)

Using the rule above and starting with 13, we generate the following sequence:

13 → 40 → 20 → 10 → 5 → 16 → 8 → 4 → 2 → 1
It can be seen that this sequence (starting at 13 and finishing at 1) contains 10 terms. Although it has not been proved yet (Collatz Problem), it is thought that all starting numbers finish at 1.

Which starting number, under one million, produces the longest chain?


```

def collatz(n):
  if(n % 2 == 0):
    return n//2
  else:
    return 3*n + 1

def collatz_seq(number):
  seq = []
  while(number!=1):
    seq.append(number)
    number = collatz(number)
  seq.append(1)
  return seq

max_len = 0
sol = []
for i in range(1,1000000):
  seq_len = len(collatz_seq(i))
  if( seq_len > max_len):
    max_len = seq_len
    sol.append(i)

print(max_len)
print(sol[-1])

    
```

**Problem 15**


---
Starting in the top left corner of a 2×2 grid, and only being able to move to the right and down, there are exactly 6 routes to the bottom right corner.


How many such routes are there through a 20×20 grid?


```
# Solved through dynamic programming by matrix filling method
# Can Also be solved by Combinatorial Methods by just finding 40 C 20 i.e 2N C N

def grid_routes(n):
  sol = []
  for i in range(0,n):
    row = []
    for i in range(0,n):
      row.append(0)
    sol.append(row)
  for i in range(0,n-1):
    sol[n-1][i] = 1
    sol[i][n-1] = 1
  for i in range(n-2,-1,-1):
    for j in range(n-2,-1,-1):
      sol[i][j] = sol[i+1][j] + sol[i][j+1]
  return sol[0][0]

grid_routes(21)
```

**Problem 50**


---
he prime 41, can be written as the sum of six consecutive primes:

41 = 2 + 3 + 5 + 7 + 11 + 13
This is the longest sum of consecutive primes that adds to a prime below one-hundred.

The longest sum of consecutive primes below one-thousand that adds to a prime, contains 21 terms, and is equal to 953.

Which prime, below one-million, can be written as the sum of the most consecutive primes?


```
from sympy import isprime
primes = []
for i in range(0,1000000):
  if(isprime(i)==True):
    primes.append(i)

max_sum = 0
maxlen = -1

for i in range(0,len(primes)):
  summ = 0
  for j in range(i,len(primes)):
    summ = summ + primes[j]
    if(summ > 1000000):
      break
    if(isprime(summ)==True and summ > max_sum and j-i > maxlen):
      maxlen = j - i +1
      max_sum = summ

print(max_sum)
print(maxlen)
```
